```python
%load_ext autoreload
%autoreload 2

from scipy.ndimage import gaussian_filter
import numpy as np

import tensorflow as tf
import matplotlib.pyplot as plt

import tensortools.numpy.images as np_img
```

```python
sz = 100
x = np.arange(-sz//2, sz//2+1)
X, Y = np.meshgrid(x, x)
R = np.sqrt(np.square(X) + np.square(Y))

z_np = np.exp(-np.square(R)/20)
print(z_np.shape)

z2_np = np_img.resize_2d(z_np, (150,150))
z3_np = np_img.resize_2d(z_np, (50,50))

fig, axs = plt.subplots(1, 3, figsize=(10,3))
pl0 = axs[0].imshow(np.abs(z_np))
pl1 = axs[1].imshow(np.abs(z2_np))
pl2 = axs[2].imshow(np.abs(z3_np))
_ = [plt.colorbar(pl, ax=axs[i]) for i, pl in enumerate([pl0, pl1, pl2])]
```
